```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("book.csv")
data.head()
```

```python
# Convert data to a binary DataFrame
data = data.applymap(bool)

# Set the minimum support and confidence values
min_support = 0.1
min_confidence = 0.5

# Apply Apriori algorithm
frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
rules
```

```python
data.head()
```

```python
# Experiment with different support and confidence values
support_values = [0.1, 0.2, 0.3,0.4]
confidence_values = [0.5, 0.6, 0.7]

for support in support_values:
    for confidence in confidence_values:
        frequent_itemsets = apriori(data, min_support=support, use_colnames=True)
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
        print(f"Support: {support}, Confidence: {confidence}, Number of Rules: {len(rules)}")
```

here we choose minimum confidence =0.6 and minimum support =0.1


```python
# Experiment with different minimum lengths
min_lengths = [1,2, 3, 4,]
frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)
for min_length in min_lengths:
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
    filtered_rules = rules[rules['antecedents'].apply(lambda x: len(x) >= min_length)]
    print(f"Minimum Length: {min_length}, Number of Rules: {len(filtered_rules)}")
```

here we choose minimum length =2

```python
# Scatter plot of support vs. confidence
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs. Confidence')
plt.show()

```

```python
# Histogram of rule lengths
sns.histplot(data=rules, x=rules['antecedents'].apply(len), bins=10)
plt.xlabel('Rule Length')
plt.ylabel('Frequency')
plt.title('Histogram of Rule Lengths')
plt.show()
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df.columns = ["V1", "V2", "V3", "V4", "V5", "Sixth Sense", "Gladiator", "LOTR1", "Harry Potter1", "Patriot", "LOTR2", "Harry Potter2", "LOTR", "Braveheart", "Green Mile"]
df = df.drop(["V1", "V2", "V3", "V4", "V5"], axis=1)
```

```python
df
```

```python
# Apply Apriori algorithm
min_support = 0.2  # Set your desired minimum support threshold
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

# Generate association rules
min_confidence = 0.7  # Set your desired minimum confidence threshold
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
rules
```

```python
support_values = [0.1, 0.15, 0.2, 0.25]
confidence_values = [0.6, 0.7, 0.8]

for support in support_values:
    for confidence in confidence_values:
        frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
        print(f"Support: {support}, Confidence: {confidence}, Number of Rules: {len(rules)}")
```

```python
# Experiment with different minimum lengths
min_lengths = [2, 3, 4,]
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
for min_length in min_lengths:
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
    filtered_rules = rules[rules['antecedents'].apply(lambda x: len(x) >= min_length)]
    print(f"Minimum Length: {min_length}, Number of Rules: {len(filtered_rules)}")
```

```python
# Scatter plot of support vs. confidence
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs. Confidence')
plt.show()
```
