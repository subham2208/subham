# case study - CUTLETS
A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units.

TEST TYPE - COMPARISION TEST ( 2SAMPLE T-TEST)

SIGNIFICANCE LEVEL (ALPHA) - 5%

DEFINING HYPOTHESIS :

NULL (H0) -**NO difference**  in the diameter of the cutlet between two units

ALTERNATE (H1) -**YES**, there is difference in the diameter of the cutlet between two units


```python
#loading data
file=files.upload()
```

```python
cutlets=pd.read_csv("Cutlets.csv")
cutlets.head()
```

```python
#Data pre-processing
cutlets.describe()
```

```python
#Data anlysis
cutlets.info()
```

```python
#since no null value hence cont testing
```

```python
#2 SAMPLE T-TEST
test=stats.ttest_ind(cutlets["Unit A"],cutlets["Unit B"])
test
```

```python
# t valu and p-value from test result tuple
print("t-value = ",test[0])
print("p-value = ",test[1])
print("----------------------------------------------")
print("since P-value > 0.05 , hence NULL IS SUPPORTED ")
print("----------------------------------------------")
```


since p-value > 0.05

Therefore , NULL HYPOTHESIS IS SUPPORTED

CONCLUDING-

1.There is no significant difference in unit-A and unit-B

2.and if there any difference , that will arise by chance or random

# case study - laboratories samples compared
whether there is any difference in average TAT among the different laboratories

TEST TYPE - COMPARISION TEST (ANOVA) since no 0f group >2

SIGNIFICANCE LEVEL (ALPHA) - 5%

DEFINING HYPOTHESIS :

NULL (H0) -**NO difference**  among the different laboratories

ALTERNATE (H1) -**YES**, there is difference among the different laboratories



```python
file=files.upload()
```

```python
labs=pd.read_csv("LabTAT.csv")
labs.head()
```

```python
#data cleaning and info
labs.describe()
```

```python
labs.info()
```

```python
#hence 4 labs data is to be compared , therefore using ANALYSIS OF VARIANCE TEST
```

```python
#USING AVONA TEST IN SCIPY STATS
test=stats.f_oneway(labs["Laboratory 1"],labs["Laboratory 2"],labs["Laboratory 3"],labs["Laboratory 4"])
```

```python
test
```

```python
#conxlude
print("f-value = ",test[0])
print("p-value = ",test[1])
print("----------------------------------------------")
print("since P-value < 0.05 , hence NULL IS REJECTED ")
print("----------------------------------------------")
```

conclusion

since p-value < 0.05

Therefore , ALTERNATIVE HYPOTHESIS IS SUPPORTED


CONCLUDING-

1.There is significant difference in average TAT among the different laboratories

2.and difference , that arise are not by chance or random

## case study - MALE-FEMALE BUYERS RATIO
Find if male-female buyer rations are similar across regions

TEST TYPE - COMPARISION TEST ( chi-square TEST for independence)

SIGNIFICANCE LEVEL (ALPHA) - 5%

DEFINING HYPOTHESIS :

NULL (H0) - male-female buyer rations are similar across regions

ALTERNATE (H1) -**YES**, there is difference in male-female buyer rations across regions


```python
#UPLOADING DATA
file=files.upload()
```

```python
#loading data
buyers=pd.read_csv("BuyerRatio.csv")
buyers
```

```python
buyer=buyers.iloc[:,1:6]
buyer
```

```python
#CHI SQUARE TEST
test=stats.chi2_contingency(buyer)
```

```python
#results
test
```

```python
print('CHI-VALUE = ',test[0])
print('P-VALUE = ',test[1])
print('---------------------------------------')
print(' since pvalue > 0.05 hence NULL HYPOTESIS SUPPORTED')
print('---------------------------------------')
```

since p-value > 0.05

Therefore , NULL HYPOTHESIS IS SUPPORTED

CONCLUDING-
** male-female buyer rations are similar across regions**

#case study -defective/error free product

The manager wants to check whether the defective %  varies by centre

TEST TYPE - COMPARISION TEST ( CHI SQUARE-TEST)

SIGNIFICANCE LEVEL (ALPHA) - 5%

DEFINING HYPOTHESIS :

NULL (H0) -**NO difference** amoung centre

ALTERNATE (H1) -**YES**, there is difference in the defective % which varies across centre

```python
#loading data
files=files.upload()
```

```python
#data manipualtion
product=pd.read_csv('Costomer+OrderForm.csv')
product.head()
```

```python
order=product.stack().reset_index()
```

```python
order.head()
```

```python
crosstab=pd.crosstab(order[0],order['level_1'])
crosstab
```

```python
#NOW CONTINGENCY TABLE IS PREPARED
#USING CHI TEST
chi,p,dof,expected_value=stats.chi2_contingency(crosstab)
```

```python
print('chi-value = ',chi)
print('p-value = ', p )
print('-------------------------------------')
print('since p-value > 0.05 hence NULL IS SUPPORTED')
print('-------------------------------------')
```

since p-value > 0.05

Therefore , NULL HYPOTHESIS IS SUPPORTED

CONCLUDING- ** defective % do not varies across centre**
